```python
import pandas as pd
df = pd.read_csv('data/debate.csv', encoding = "ISO-8859-1")
import numpy as np
import matplotlib.pyplot as plt
import nltk
%matplotlib notebook
```

As in the data story the following code organises the text from the debate speech into various lists and strings with and without stop words/punctuation.

```python
clinton_whole_text = ''
trump_whole_text = ''
c = []

for index, row in df.iterrows():
    if row[1] == 'Clinton':
        clinton_whole_text += row[2] + ' '
        c.append((row[2], 'Clinton'))
    elif row[1] == 'Trump':
        trump_whole_text += row[2] + ' '
        c.append((row[2], 'Trump'))

import string
exclude = set(string.punctuation)
clinton_whole_text_excluded = ''.join(ch for ch in clinton_whole_text if ch not in exclude)
trump_whole_text_excluded = ''.join(ch for ch in trump_whole_text if ch not in exclude)

from stop_words import get_stop_words

stop_words = get_stop_words('english')
stop_words_string = " ".join(str(x) for x in stop_words)
stop_words = ''.join(ch for ch in stop_words_string if ch not in exclude)


clinton_whole_text_excluded = [i for i in clinton_whole_text_excluded.lower().split() if i not in stop_words]
trump_whole_text_excluded = [i for i in trump_whole_text_excluded.lower().split() if i not in stop_words]
```

It's interesting that the word 'people' featured heavily in the dialogue from both Trump and Clinton. NLTK allows us to look at the context in which the word was used by both candidates. We can see that while Clinton focussed more on the economy, education and wealth inequality, Trump focussed more on international trade, terrorism, gun crime and immigration. The word 'people' is clearly an anchor word but it is being used by each candidate in very different context.

```python
clinton_token = nltk.word_tokenize(clinton_whole_text)
clinton_text = nltk.Text(clinton_token)
clinton_text.concordance('people')
```

```python
trump_token = nltk.word_tokenize(trump_whole_text)
trump_text = nltk.Text(trump_token)
trump_text.concordance('people')
```

Perhaps a Deep Learning Neural Network with classifiers could be used to recognise and generate text from each candidate. Below is an attempt to begin to build a rudimentary predictive model of each candidate's speech.

```python
word_frequency_clinton = {}
word_frequency_trump = {}
word_frequency_total = {}
whole_text_excluded = clinton_whole_text_excluded + trump_whole_text_excluded
for word in clinton_whole_text_excluded:
        if word in word_frequency_clinton.keys():
            word_frequency_clinton[word] +=1
        else:
            word_frequency_clinton[word] = 1
for word in trump_whole_text_excluded:
        if word in word_frequency_trump.keys():
            word_frequency_trump[word] +=1
        else:
            word_frequency_trump[word] = 1
for word in whole_text_excluded:
        if word in word_frequency_total.keys():
            word_frequency_total[word] +=1
        else:
            word_frequency_total[word] = 1
```

```python
word_frequency_total
```

We can see below that Trump has many highly predictive words. Clinton less so. Perhaps being able to dominate the word space and lay claim to this language is what helped Trump to win? 

```python
word_ratio_trump = {}


for word in word_frequency_trump:
    if word_frequency_trump[word] > 20:
        if word_frequency_trump[word]/word_frequency_total[word] > 0.7:
            word_ratio_trump[word]=word_frequency_trump[word]/word_frequency_total[word]

word_ratio_trump
```

```python
word_ratio_clinton = {}

for word in word_frequency_clinton:
        if word_frequency_clinton[word] > 20:
            if word_frequency_clinton[word]/word_frequency_total[word] > 0.7:
                word_ratio_clinton[word]=word_frequency_clinton[word]/word_frequency_total[word]
word_ratio_clinton
```

This brief foray into NLP has yielded greater predictive insight into the language used by the candidates. We can clearly see that while Trump and Clinton use similar words, the context of those words is very different, reflecting a different perspective of the targetted demographics.
